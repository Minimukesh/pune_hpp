## Two Data Structures

```python
1. DataFrame:
    2D Data Structure >> Multiple Columns
    It consist of multiple 1D 
    Table with rows and columns
    
2. Series:
    1D Data Structure >> Single Column
    like an array
```

```python
1. Read Data from (CSV, EXcel, JSON)
2. Read Data from DataBase (SQL and NoSQL)
3. Exploratory Data Analysis >> EDA
4. Data PreprocessingCleaning:
    Missing Values
    Outliers Handling
    
5. Encoding:
    1. Label Encoder
    2. One Hot Encoder
    
6. Data Preprocessing/Data Manipulation/Data Wrangling/Data Munging >> Feature Engineering
7. Feature Selection/Importance
8. JOinng Multiple Tables (left, right, inner, outer)
9. Provides Time Series Functionality

Columns >> Features/Variables/Attributes/ Parameteres/Columns
Rows >> Entries/ Datapoints/ Observation/Records
```
